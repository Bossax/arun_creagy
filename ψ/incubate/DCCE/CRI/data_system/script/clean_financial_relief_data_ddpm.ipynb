```python
import pandas as pd
import numpy as np
```

This is a notebook to extract nested excel sheets into raw data tables

```python
file_path = 'สถิติข้อมูลการใช้จ่ายเงินทดรองราชการ ปี 2546 - ปัจจุบัน.xlsx'
xl = pd.ExcelFile(file_path)
```

### Cleaning Yearly Sheets: Per hazard 

We start from sheet 5 (index 4) up to sheet 19 (index 18).
Each sheet has custom slicing applied based on the table layout.

```python
# Define the correct column names for the disaster sheets
table_columns = [
    'จังหวัด', 'อุทกภัย', 'ภัยแล้ง', 'ฝนทิ้งช่วง', 'ภัยหนาว', 
    'อัคคีภัย', 'วาตภัย', 'โรคศัตรูพืช', 'โรคระบาดสัตว์', 'ภัยอื่น', 
    'รวมทั้งสิ้น', 'วงเงินเสนอ', 'วงเงินอนุมัติ', 'วงเงินไม่อนุมัติ'
]

cleaned_dataframes = {}

# Process sheets from position 5 to 19 (indices 4 to 18)
for sheet_name in xl.sheet_names[4:19]:
    # Read the sheet without headers initially
    df = xl.parse(sheet_name, header=None)
    
    # Cut rows and columns based on your specific slicing
    df_cleaned = df.iloc[5:87, 2:16].copy()
    
    # Assign clean headers
    df_cleaned.columns = table_columns
    
    # Clean up: remove rows where 'จังหวัด' is missing or is 'หมายเหตุ'
    df_cleaned = df_cleaned.dropna(subset=['จังหวัด'])
    df_cleaned = df_cleaned[df_cleaned['จังหวัด'] != 'หมายเหตุ']
    
    # Add a 'ปี' column (numeric BE) as the second column
    import re
    year_match = re.search(r'(\d+)', sheet_name)
    if year_match:
        df_cleaned.insert(1, 'ปี', int(year_match.group(1)) + 2500)
    
    cleaned_dataframes[sheet_name] = df_cleaned.reset_index(drop=True).fillna(0)

print(f"Processed {len(cleaned_dataframes)} hazard sheets (4:19).")
```

#### Hazard sheets: Position 19 to 28 (Older records)


```python
# Define the correct column names for older disaster sheets
table_columns2 = [
    'จังหวัด', 'อุทกภัย', 'ภัยแล้ง', 'ฝนทิ้งช่วง', 'ภัยหนาว', 
    'อัคคีภัย', 'วาตภัย', 'โรคศัตรูพืช', 'โรคระบาดสัตว์', 'ภัยอื่น', 
    'รวมทั้งสิ้น'
]

# Process sheets from position 19 to 28
for sheet_name in xl.sheet_names[19:28]:
    df = xl.parse(sheet_name, header=None)
    df_cleaned = df.iloc[5:87, 2:13].copy()
    df_cleaned.columns = table_columns2
    
    df_cleaned = df_cleaned.dropna(subset=['จังหวัด'])
    df_cleaned = df_cleaned[df_cleaned['จังหวัด'] != 'หมายเหตุ']
    
    # Add a 'ปี' column (numeric BE) as the second column
    import re
    year_match = re.search(r'(\d+)', sheet_name)
    if year_match:
        df_cleaned.insert(1, 'ปี', int(year_match.group(1)) + 2500)
    
    cleaned_dataframes[sheet_name] = df_cleaned.reset_index(drop=True).fillna(0)

print(f"Processed {len(cleaned_dataframes)} total hazard sheets.")
```

### Cleaning Yearly Sheets: Per sector 


```python
# Define the correct column names for the sector sheets
table_columns_sector = [
    'จังหวัด', 'ด้านดำรงชีพ', 'ด้านสังคมสงเคราะห์', 'ด้านการแพทย์และสาธารณสุข', 
    'ด้านเกษตร_พืช', 'ด้านเกษตร_ประมง', 'ด้านเกษตร_ปศุสัตว์', 'ด้านเกษตร_อื่น', 
    'ด้านบรรเทาสาธารณภัย', 'ด้านการปฏิบัติงานบรรเทาทุกข์', 'เชิงป้องกันหรือยับยั้ง', 
    'รวมทั้งสิ้น', 'วงเงินเสนอ', 'วงเงินอนุมัติ', 'วงเงินไม่อนุมัติ'
]

cleaned_dataframes_sector = {}

# Process sector sheets from '68ด้าน' to '55ด้าน' (indices 28 to 42)
for sheet_name in xl.sheet_names[28:43]:
    df = xl.parse(sheet_name, header=None)
    df_cleaned = df.iloc[5:87, 2:17].copy()
    df_cleaned.columns = table_columns_sector
    
    df_cleaned = df_cleaned.dropna(subset=['จังหวัด'])
    df_cleaned = df_cleaned[df_cleaned['จังหวัด'] != 'หมายเหตุ']

    # Add a 'ปี' column (numeric BE) as the second column
    import re
    year_match = re.search(r'(\d+)', sheet_name)
    if year_match:
        df_cleaned.insert(1, 'ปี', int(year_match.group(1)) + 2500)
    
    cleaned_dataframes_sector[sheet_name] = df_cleaned.reset_index(drop=True).fillna(0)

print(f"Processed {len(cleaned_dataframes_sector)} sector sheets (28:43).")
```

```python
# Define the correct column names for older sector sheets
table_columns_sector2 = [
    'จังหวัด', 'ด้านดำรงชีพ', 'ด้านสังคมสงเคราะห์', 'ด้านการแพทย์และสาธารณสุข', 
    'ด้านเกษตร_พืช', 'ด้านเกษตร_ประมง', 'ด้านเกษตร_ปศุสัตว์', 'ด้านเกษตร_อื่น', 
    'ด้านบรรเทาสาธารณภัย', 'ด้านการปฏิบัติงานบรรเทาทุกข์', 'เชิงป้องกันหรือยับยั้ง', 
    'รวมทั้งสิ้น'
]

# Process sector sheets from '54ด้าน' to '46ด้าน' (indices 43 to 52)
for sheet_name in xl.sheet_names[43:52]:
    df = xl.parse(sheet_name, header=None)
    df_cleaned = df.iloc[5:87, 2:14].copy()
    df_cleaned.columns = table_columns_sector2
    
    df_cleaned = df_cleaned.dropna(subset=['จังหวัด'])
    df_cleaned = df_cleaned[df_cleaned['จังหวัด'] != 'หมายเหตุ']

    # Add a 'ปี' column (numeric BE) as the second column
    import re
    year_match = re.search(r'(\d+)', sheet_name)
    if year_match:
        df_cleaned.insert(1, 'ปี', int(year_match.group(1)) + 2500)
    
    cleaned_dataframes_sector[sheet_name] = df_cleaned.reset_index(drop=True).fillna(0)

print(f"Processed {len(cleaned_dataframes_sector)} total sector sheets.")
```

### Consolidating Master Database

We concatenate all sheets and fill missing columns with 0.

```python
# 1. Consolidate Disaster Sheets
master_disaster_df = pd.concat(cleaned_dataframes.values(), ignore_index=True)
# Fill NaNs with 0 (for older sheets missing newer columns)
master_disaster_df = master_disaster_df.fillna(0)

# 2. Consolidate Sector Sheets
master_sector_df = pd.concat(cleaned_dataframes_sector.values(), ignore_index=True)
# Fill NaNs with 0
master_sector_df = master_sector_df.fillna(0)

# 3. Save to CSV
master_disaster_df.to_csv('master_financial_relief_by_hazard.csv', index=False, encoding='utf-8-sig')
master_sector_df.to_csv('master_financial_relief_by_sector.csv', index=False, encoding='utf-8-sig')

print("Master database files created: 'master_financial_relief_by_hazard.csv' and master_financial_relief_by_sector.csv'")
print(f"Disaster records: {len(master_disaster_df)}")
print(f"Sector records: {len(master_sector_df)}")
```
